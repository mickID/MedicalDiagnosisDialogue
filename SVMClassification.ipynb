```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import svm
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from yellowbrick.features import Manifold
```

```python
# Read the data
data = pd.read_csv('ExpImpDev.csv')

diseaseList = np.empty((0,12))
with open('disease.txt') as f:
    lines = f.readlines()
    for el in lines:
      el = el.replace('\n', '')
      diseaseList = np.append(diseaseList, el)
print(diseaseList)
```

```python
X = data.to_numpy()
y = X[:, 118]
X = X[:, :118]

#Shuffling samples
rng = np.random.RandomState(42)
p = rng.permutation(len(y))
X, y = X[p, :], y[p]
```

```python
#Test Polynomial kernel with cross validation
clf = svm.SVC(kernel='poly', degree=2, C=50, decision_function_shape='ovo')
y_pred_score = cross_val_score(clf, X, y, cv=5)
print(y_pred_score)
```

```python
#Finding best parameters with Gridsearch
param = {
    'kernel':('poly', 'rbf'),
    'C':[1, 3, 5, 10, 25, 50],
    'degree':[2, 3, 4, 5],
    'coef0':[0.5, 1, 5, 7, 10],
    'gamma':('auto', 'scale'),
    'decision_function_shape':['ovo']
}
clf = svm.SVC()
grd = GridSearchCV(clf, param, cv=5)
grd.fit(X, y)
grd.best_params_
```

```python
#Test Polynomial kernel with cross validation
clf = svm.SVC(kernel='poly', degree=3, C=1, gamma='scale', coef0=5, decision_function_shape='ovo')
y_pred_score = cross_val_score(clf, X, y, cv=5)
print(y_pred_score)
```
