```python
import pandas as pd
import numpy as np
import pickle
```

```python
goals = {}
with open('train.pk', 'rb') as f:
  goals['train'] = pickle.load(f)
train = goals['train']
print(train)
```

```python
sympthoms = np.empty((0,120))
sympthoms = np.append(sympthoms, 'key')
with open('symptom.txt') as f:
    lines = f.readlines()
    for el in lines:
      el = el.replace('\n', '')
      sympthoms = np.append(sympthoms, el)
sympthoms = np.append(sympthoms, 'disease_tag')

df = pd.DataFrame(columns=sympthoms)
```

```python
x = np.zeros((1, sympthoms.size), dtype=object)
i = 0
for sample in train:
  x[i, 0] = i
  disease_tag = sample['disease_tag']
  data = pd.DataFrame(x, columns=sympthoms)
  data.at[i, 'disease_tag'] = disease_tag
  # Considero i sintomi espliciti
  for sympthom in sample['explicit_inform_slots']:
    value = sample['explicit_inform_slots'][sympthom]
    if(value):
      data.at[i, sympthom] = 1;
  # Considero i sintomi impliciti
  for sympthom in sample['implicit_inform_slots']:
    value = sample['implicit_inform_slots'][sympthom]
    if(value):
      data.at[i, sympthom] = 1;
  x = data.to_numpy()
  z = np.zeros((1, sympthoms.size), dtype=object)
  x = np.vstack((x, z))
  i = i+1
```
